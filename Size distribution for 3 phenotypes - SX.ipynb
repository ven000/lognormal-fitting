```python
import pandas as pd
%matplotlib inline 
import numpy as np
from scipy.stats import lognorm, norm
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import scipy.stats as stats


path = r'~/Downloads/Book10.csv'

df=pd.read_table(path)
features = df.columns[0].split(',')
df2=df[df.columns[0]].str.split(',', expand=True)
# df1.columns = df1.iloc[0]
df2.columns = features
df2=df2.drop(['Count Type','Positivity in Channel 1','Positivity in Channel 2','ROI Name','Channel Identifier','Raw Channel Identifier','export id','app settings','dataset','dataset id'], axis=1)


df2 = df2[df2.columns].astype(float)

rslt_df_1 = df2[(df2['Counts in Channel 1'] >= 3 )& (df2['Counts in Channel 2'] >= 3)]  #channel 1&2 positive
radius = rslt_df_1[rslt_df_1['Radius of gyration (nm)']<300]['Radius of gyration (nm)']
# radius = rslt_df['radius_of_gyration']
print('the number of double positive particles is: ')
print(len(rslt_df_1))
print('median radius is: ')
print(radius.median())
chan_1_2 = len(radius)

plt.hist(radius,bins=50)
plt.show()
```
